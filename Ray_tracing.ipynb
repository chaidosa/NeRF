```python
H = 400 
W = 400
# H and W represents the size of the image

f = 1200 # distance between the pinhole camera and the plane where the image will be formed
```

```python
rays_o   = np.zeros((H*W, 3)) #rays origin
rays_dir = np.zeros((H*W, 3)) #rays direction

u = np.arange(W)
v = np.arange(H)

u, v = np.meshgrid(u, v)

dirs = np.stack((u - W / 2,
                 -(v - H) /2,
                 -np.ones_like(u) * f), axis = -1)

rays_d = dirs / np.linalg.norm(dirs, axis= -1, keepdims=True)
```
